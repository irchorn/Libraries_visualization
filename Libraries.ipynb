```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import warnings 
warnings.filterwarnings("ignore")
```

```python
#Import dataset
libraries_file = 'NYPL_Branch_Services_-_Manhattan.csv'
df_libraries = pd.read_csv(libraries_file)
display(df_libraries.head())
```

```python
df_libraries.shape
```

```python
df_libraries.info()
```

```python
columns = list(df_libraries.columns)
print(columns)
```

```python
#Relationship between ADULT Program and ADULT Attendance

plt.scatter(df_libraries['ADULT Program'], df_libraries['ADULT Attendance'], 
            c = df_libraries['Weekly Hours of Public Service'], cmap = 'winter')

plt.xlabel('ADULT Program')
plt.ylabel('ADULT Attendance')
cbar= plt.colorbar()
cbar.set_label('Weekly Hours')
```

```python
#Correlation heatmap
plt.figure(figsize = (10,8))
sns.heatmap(df_libraries.corr(method='spearman'), cmap = 'coolwarm')
plt.show()
```

Adult attendance has high correlation with adult program.

```python
#Sorting library branches by the number of adult programs from high to low
df_top_branches = df_libraries.sort_values('ADULT Program' , ascending=False)
```

```python
df_top_branches
```

```python
#Visualizing mean Adult Attendance of Adult Program at top 5 branches
g = sns.catplot(x="ADULT Program", y="ADULT Attendance", col="Branch",
                data=df_top_branches[3:8], kind="bar",
                height=4, aspect=.7);
```

58th Street branch has the highest Adult Attendance.

```python
#Pie Chart
plt.figure(figsize = (8,6))
slices = [6095, 120, 743]
labels = ['ADULT Attendance', 'JUVENILE Attendance', 'OUTREACH SERVICES Attendance']
explode = [0, 0, 0.1]
plt.pie(slices, labels = labels, explode = explode, autopct = '%1.1f%%', wedgeprops ={'edgecolor':'black'})
plt.title("58th Street Program Attendance")
plt.tight_layout()
plt.show()
```

Adult attendance at 58th street branch has the highest percentage.
